```python
import csv
from youtube_transcript_api import YouTubeTranscriptApi

# YouTube video URL (The launch event youtube link)
video_url = "https://www.youtube.com/watch?v=pxlaUCJZ27E"

# Define the output CSV file name
output_csv = "google_lauch.csv"

def download_and_save_transcript(video_url, output_file):
    try:
        video_id = video_url.split("v=")[1]
        transcript = YouTubeTranscriptApi.get_transcript(video_id)
        
        with open(output_file, 'w', newline='', encoding='utf-8') as csvfile:
            fieldnames = ['text', 'start', 'duration']
            writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
            writer.writeheader()
            
            for entry in transcript:
                writer.writerow({'text': entry['text'], 'start': entry['start'], 'duration': entry['duration']})
        
        print(f"Transcript saved to {output_file}")
    except Exception as e:
        print(f"An error occurred: {str(e)}")

if __name__ == "__main__":
    download_and_save_transcript(video_url, output_csv)
```

```python


import nltk
from nltk.corpus import stopwords

nltk.download('stopwords') #removing stopwords from the text column
```

```python
# Define the name of the column to process
column_name = "text"

# Function to remove stop words from a text column
def remove_stop_words(text):
    stop_words = set(stopwords.words("english"))
    words = text.split()  # Split the text into words
    filtered_text = [word for word in words if word.lower() not in stop_words]
    return " ".join(filtered_text)

# Apply the remove_stop_words function to the specified column
gg[column_name] = gg[column_name].apply(remove_stop_words)
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
```

```python
text_column_name = "text"  

# Combine all the text from the specified column into a single string
text = ' '.join(gg[text_column_name])

# Function to remove stop words from the combined text
def remove_stop_words(text):
    stop_words = set(stopwords.words("english"))
    words = text.split()
    filtered_text = [word for word in words if word.lower() not in stop_words]
    return " ".join(filtered_text)

# Remove stop words from the text
filtered_text = remove_stop_words(text)

# Create a WordCloud object
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(filtered_text)

# Display the word cloud using Matplotlib
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")  
plt.show()
```
